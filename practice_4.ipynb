```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cat4=bank4.select_dtypes(include="object").columns
num4=bank4.select_dtypes(exclude="object").columns
```

```python
for i in range(0,len(cat4)):
    values=bank4[cat4[i]].value_counts().values
    keys=bank4[cat4[i]].value_counts().keys()
    bankdf4=pd.DataFrame(zip(keys,values),columns=[cat4[i],'values'])
    print(f"...............{cat4[i]}..........")
    print(bankdf4)
```

```python
for i in range(1,3):
    value4=bank4[cat4[i]].value_counts().values
    keys4=bank4[cat4[i]].value_counts().keys()
    bankdf4=pd.DataFrame(zip(keys4,value4),columns=[cat4[i],'values'])
    print(f"...................{cat4[i]}...............")
    print(bankdf4)
```

```python
plt.figure(figsize=(10,8))
colors4=['red','green','yellow','black']
for i in range(1,3):
    values4=bank4[cat4[i]].value_counts().values
    keys4=bank4[cat4[i]].value_counts().keys()
    plt.subplot(2,2,i).bar(keys4,values4,color=colors4)
    plt.xlabel(cat4[i])
    plt.ylabel("frequency")
    plt.title("Bar_Chart")
    plt.savefig("marital.jpg")
```

```python
plt.figure(figsize=(10,5))
for i in range(1,3):
    keys4=bank4[cat4[i]].value_counts().keys()
    values4=bank4[cat4[i]].value_counts().values
    values_uni4=bank4[cat4[i]].unique()
    plt.subplot(2,2,i).pie(values4,labels=keys4,autopct='%0.2f%%',
                          explode=[0.1 for i in range(len(values_uni4))],
                          shadow=True,
                          startangle=180)
    
```
