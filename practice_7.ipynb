```python
cat7=bank7.select_dtypes(include="object").columns
num7=bank7.select_dtypes(exclude="object").columns
```

```python
for i in range(1,3):
    values=bank7[cat7[i]].value_counts().values
    keys=bank7[cat7[i]].value_counts().keys()
    bankdf7=pd.DataFrame(keys,values,columns=[cat7[i],'values'])
    print(f"............{cat7[i]}..................")
    print(bankdf7)
```

```python
plt.figure(figsize=(10,5))
colors=['red','green','blue','yellow']
for i in range(1,3):
    values=bank7[cat7[i]].value_counts().values
    keys=bank7[cat7[i]].value_counts().keys()
    plt.subplot(1,2,i).bar(keys,values,color=colors)
    plt.xlabel(cat7[i])
    plt.ylabel("frequency")
    plt.title("bar_chart")
    plt.savefig("continent.jpg")
    
```

```python
plt.figure(figsize=(10,5))
for i in range(1,3):
    value7=bank7[cat7[i]].value_counts().values
    keys7=bank7[cat7[i]].value_counts().keys()
    uni_keys=bank7[cat7[i]].value_counts().unique()
    plt.subplot(2,2,i).pie(value7,labels=keys7,autopct='%0.2f%%',
                          explode=[0.05 for i in range(len(uni_keys))],
                          shadow=True,
                          startangle=180)
    
```
